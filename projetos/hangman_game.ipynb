```python
import random

def get_word():
    word_list = ['gato', 'cachorro', 'lagarto', 'centop√©ia', 'boi']
    word = random.choice(word_list)
    return word.upper()
```

```python
def play(word):
    word_completion = '_' * len(word)
    guessed_letters = []
    guessed_words = []
    tries = 6
    guessed = False
    
    print(f'JOGO DA FORCA!\nDica: {len(word)} letras!')
    print(display_hangman(tries))
    print(word_completion, '\n')
    
    while not guessed and tries > 0:
        guess = input('Advinhe uma letra ou palavra: ').upper()
        
        if len(guess) == 1 and guess.isalpha():
            if guess in guessed_letters:
                print(f'Voc√™ j√° tentou a letra {guess}')
            
            elif guess not in word:
                print(f'{guess}, N√£o!')
                tries -= 1
                guessed_letters.append(guess)
            
            else:
                print(f'{guess} est√° na palavra!')
                guessed_letters.append(guess)
                word_as_list = list(word_completion)
                
                index = [i for i, letter in enumerate(word) if letter == guess]
                
                for idx in index:
                    word_as_list[idx] = guess
                word_completion = ''.join(word_as_list)
                
                if '_' not in word_completion:
                    guessed = True
        
        elif len(guess) == len(word) and guess.isalpha():
            if guess in guessed_words:
                print(f'Voc√™ j√° tentou a palavra {guess}')
            
            elif guess != word:
                print(f'{guess} n√£o √© a palavra!')
                tries -= 1
                guessed_words.append(guess)
            
            else:
                guessed = True
                word_completation = word       
        
        else:
            print('Inv√°lido!')
        print(display_hangman(tries))
        print(word_completion, '\n')
    
    if guessed:
        print('Parab√©ns, voc√™ acertou a palavra! Voc√™ venceu!')
    
    else:
        print(f'Voc√™ perdeu! A palavra era: {word}.')    
```

```python
def display_hangman(tries):
    stages = [  # cabe√ßa, tronco, bra√ßos e pernas: morte
                """
                   --------
                   |      |
                   |      O
                   |     \|/
                   |      |
                   |     / \
                   -
                """,
                # cabe√ßa, tronco e bra√ßos, perna
                """
                   --------
                   |      |
                   |      O
                   |     \|/
                   |      |
                   |     / 
                   -
                """,
                # cabe√ßa, tronco e bra√ßos
                """
                   --------
                   |      |
                   |      O
                   |     \|/
                   |      |
                   |      
                   -
                """,
                # cabe√ßa, tronco e bra√ßo
                """
                   --------
                   |      |
                   |      O
                   |     \|
                   |      |
                   |     
                   -
                """,
                # cabe√ßa e tronco
                """
                   --------
                   |      |
                   |      O
                   |      |
                   |      |
                   |     
                   -
                """,
                # cabe√ßa
                """
                   --------
                   |      |
                   |      O
                   |    
                   |      
                   |     
                   -
                """,
                # vazio
                """
                   --------
                   |      |
                   |      
                   |    
                   |      
                   |     
                   -
                """
    ]
    return stages[tries]
```

```python
def main():
    word = get_word()
    play(word)
    while input('Jogar de novo? (S/N) ').upper() == 'S':
        word = get_word()
        play(word)
        
        
if __name__ == '__main__':
    main()
```
