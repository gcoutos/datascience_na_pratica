```python
from chatterbot import ChatBot
from chatterbot.trainers import ListTrainer
from chatterbot.trainers import ChatterBotCorpusTrainer

bot = ChatBot('Python')
bot = ChatBot(
    'Python',
    storage_adapter='chatterbot.storage.SQLStorageAdapter',
    database_uri='sqlite:///database.sqlite3',
    logic_adapters=["chatterbot.logic.BestMatch"]
    )

with open('talk_bot.txt', 'r', encoding='utf-8') as open_file:
    words_list = open_file.read()
    words_list.split(', ')

talk = ListTrainer(bot)
talk.train(words_list)

trainer = ChatterBotCorpusTrainer(bot)
trainer.train('chatterbot.corpus.portuguese')
```

```python
while True:
    answer = bot.get_response(input('Usu√°rio: '))
    if float(answer.confidence) > 0.5:
        print(f'PyBot: {answer}')
    else:
        print('Desculpe, n√£o entendi!')
        break
```
